## Feature Engineering
In the above dataset except price everything is object
so we have to convert it into numerical value

### working with Date_of_Journey

```python
# Separating object values into different position
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
# Every value except price is object so we have to convert them into numerical value
```

```python
# Converting object value into integer value
df['Date']= df['Date'].astype(int)
df['Month']= df['Month'].astype(int)
df['Year']= df['Year'].astype(int)
```

```python
df.info()
```

```python
# Now drop Date_of_Journey column because it will not add value
df.drop('Date_of_Journey', axis=1, inplace=True)
df.head()
```

### Working with arrival

```python
# Arrival_time: split data into hour and minutes first
# Extract the hour from the Arrival_Time column
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
# Extract the minute from the Arrival_Time column, handling potential extra information
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1].str.split(' ').str[0]# first split in ":" and take 2nd value ,then split in space and take 1st value 2:20 March, fist it will take "20 March" then it will take only "20" out of it
```

```python
df.head()
```

```python
df.info()
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']= df['Arrival_min'].astype(int)
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
df.head()
```

### Working with departure time

```python
# Dep_Time: split data into hour and minutes first
# Extract the hour from the Arrival_Time column
df['Dep_hour']=df['Dep_Time'].str.split(':').str[0]
# Extract the minute from the Arrival_Time column, handling potential extra information
df['Dep_min']=df['Dep_Time'].str.split(':').str[1]
df.head()
```

```python
df.info()
```

```python
df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_min']= df['Dep_min'].astype(int)
df.info()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
df.head()
```

### Working with duration

```python
df['Duration_hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df.head()
```

```python
df['Duration_hour'].unique()
```

```python
# 5min = 0.08h
df['Duration_hour']=df['Duration_hour'].replace('5m','0.08')
```

```python
df['Duration_min'].unique()
```

```python
df['Duration_min'].isnull().sum()
```

```python
df['Duration_min'] = df['Duration_min'].fillna(0)
```

```python
df['Duration_min'].isnull().sum()
```

```python
df.info()
```

```python
df['Duration_hour']=df['Duration_hour'].astype(float)
df['Duration_min']=df['Duration_min'].astype(int)
df.info()
```

```python
df.drop('Duration',axis=1,inplace=True)
df.head()
```

### Working with total stops

```python
df['Total_Stops'].unique()
```

- this is categorical data so we are taking mode to fill nan data

```python
df['Total_Stops'].mode()
```

```python
#target ordinal encoding
encoded_in_order = {'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4, np.nan:1}
df['Total_Stops']=df['Total_Stops'].map(encoded_in_order)
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.drop('Route',axis=1,inplace=True)
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder= OneHotEncoder()
encoded = encoder.fit_transform(df[['Airline', 'Source', 'Destination', 'Additional_Info']]).toarray()
encoded
```

```python
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
encoded_df.head()
```

```python
df= pd.concat([df, encoded_df], axis=1)
df.head()
```

```python
df.drop(['Airline', 'Source', 'Destination', 'Additional_Info'], axis=1, inplace=True)
df.head()
```
